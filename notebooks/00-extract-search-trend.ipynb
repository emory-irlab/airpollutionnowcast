```python
def extract_search_trend(config_path, output_filepath):
    """ Runs data processing scripts to turn raw data from (../raw) into
        cleaned data ready to be analyzed (saved in ../interim).
    """
#     logger = logging.getLogger(__name__)
#     logger.info('making final data set from raw data')

    pars = configparser.ConfigParser()
    pars.read(config_path)
    city = pars['global']['city']
    seed_path = pars['extract_search_trend']['term_list_path']
    search_data_path = pars['extract_search_trend']['search_data_path']
    name_pattern = pars['extract_search_trend']['name_pattern']
    train_years = ast.literal_eval(pars['global']['train_year'])
    test_years = ast.literal_eval(pars['global']['test_year'])
    season = pars['global']['season']

    raw_file_path = generate_file_path(search_data_path, name_pattern, city)
    trend_data = read_raw_data(raw_file_path)
    years = train_years + test_years
    return years, trend_data
    trend_data = select_years(trend_data, years)
    trend_data = select_season(trend_data, season)
    # Get seed_word_list
    seed_word_list = [k.lower() for k in pd.read_csv(seed_path, header=None)[0].values]
    trend_data = select_terms(trend_data, seed_word_list)
    trend_data.to_csv(output_filepath)
```

```python
config_path = 'config/parameters.ini'
output_filepath = 'data/interim/search.csv'
```
